The handy model uses a set of differential equations (1-4) to define human and nature dynamics. Eq. 5-6 are how the consumptions and death rate of commoners and elites in the model
change if there is famine in the world. Handy specifies that there is famine in the world if there is wealth less than wealth threshold. The wealth threshold is defined by minimum wealth required by the commoners and elites to survive.
$$ 
x'_{c} = \beta_{c} x_{c} - \alpha_{c} x_{c}                           --------(1)
\\
x'_{e} = \beta_{e} x_{e} - \alpha_{e} x_{e}                           --------(2)
\\
y' = \gamma y (\lambda - y) - \delta x_{c} y                          --------(3)
\\
w' = \delta x_{c} y - C_{c} - C_{E}                                   --------(4)
\\
where, 
\\
C_{c} = min(1,\frac{w}{w_{th}}) s x_{c}                                --------(5)
\\
C_{e} = min(1,\frac{w}{w_{th}}) \kappa s x_{c}                                --------(6)
$$

#### 1. Egalitarian Society: This is society with no elite population. 

1.1 Here optimal equilibrium can be achieved for a certain extraction (production factor). As proposed by the model, if the extraction rate is higher than optimal value, it leads to an oscillatory approach to a sub-optimal carrying capacity value. 
The optimal value of production factor in HANDY for the values used in modeling is: 6.67 E -6. Using this value along with no elite in the model produces Figure 3a from HANDY paper. 

![optimal](egalitarian_optimal.jpg)

1.2 Oscillatory approach to suboptimal solution can be seen in the model by increasing the value of production_factor to 9.9E-6. This produces Figure 3b from the HANDY paper. 

![suboptimal](oscillatory_suboptimal.jpg)

1.3 Cycles of prosperity, overshoot and collapses can be seen in the model when you put the value of production factor to be 9.0E-5. Thsi produces Figure 3c from HANDY paper. 

![n_cycle](N_Cyclic.jpg)

1.4 Full collapse can be seen in the model when you put the value of production factor to be 9.0E-4. Thsi produces Figure 3d from HANDY paper. 

![collapse](Full_collapse.jpg)

#### 2. Equitable Society: This is society with elite population but there is no income inquality factor (i.e. kappa = 1). 

2.1 Here optimal equilibrium can be achieved for a certain extraction (production factor). As proposed by the model, if production factor is 8.33E-6. This value is higher when there are no elites because commoners are forced to extract nature at a higher rate for increased burden of elites in the population. Using this value along with 25 elites in the model produces Figure 4a from HANDY paper. 

![equitableoptimal](equitable_optimal.jpg)

2.2 Oscillation_suboptimal value can be achieved in the equitable society by setting the value to 2.20E-5 and number of elites to be 25. We get the Figure 4b from the HANDY paper. 

![equitable_suboptimal](equitable_oscillation.jpg)

2.3 N_cyclic phenomenon in equitable society value can be achieved by setting the value to 3.0E-5 and number of elites to be 25. We get the Figure 4c from the HANDY paper. 

![equitable_cycles](N_cyclic_equitable.jpg)

#### 3. Unequal Society: This is society with elite population and there is income inquality factor (i.e. kappa > 1). 

3.1 The L collapse of population where elite and commoners die but nature recoveres after sometime can be reproduced from the code using a value of 6.6E-5 (in HANDY paper they suggest using a value of 6.6E-6). That vale also produces an L collapse but it does not look anything like Figure 6a in the paper.

```python
############################## IMPORTING THE RIGHT LIBRARIES ########################################

import numpy as np                       # for mathematical functions
import matplotlib.pyplot as plt          #for plotting the results

#####################################################################################################

############################## INITIALIZING CONSTANTS ###############################################

nominal_death_rate = 1E-2
famine_death_rate = 7E-2
commoners_birth_rate = 3E-2
elites_birth_rate = 3E-2
subsistence_salary = 5E-4
threshold_wealth_pc = 5E-3
regeneration_rate = 1E-2
nature_carrying_capacity = 1E+2
inequality_factor = 1   # we try 1, 10 and 100

#####################################################################################################

############################# INITIALIZING VARIABLES ###############################################
time_step = 0.1
start_time = 0
end_time = 1000
time  = np.arange(start_time, end_time, time_step)
time = np.reshape(time, (-1,1))
commoner_population = np.empty([len(time),1])
commoner_population[0] = 100
elite_population = np.empty([len(time),1])
elite_population[0] = 0
nature = np.empty([len(time),1], dtype = "float")
nature[0] = nature_carrying_capacity
accum_wealth = np.empty([len(time),1])
accum_wealth[0] = 0
famine_ratio = np.empty([len(time),1])
famine_ratio[0] = 0
wealth_threshold = np.empty([len(time),1])
wealth_threshold[0] = 0
#controls of simulation
production_factor = 6.67E-6
pop_dim = (famine_death_rate - commoners_birth_rate) / (famine_death_rate - nominal_death_rate)
#print(pop_dim)

############################# CODE FOR THE SIMULATION ####################################
# Please do not chnage this section unless you are confident in what you are doing. #

def HANDY(time):
    for i in range (0, len(time)-1):
    #updating the varibale
        wealth_threshold[i] = threshold_wealth_pc * commoner_population[i] + inequality_factor * threshold_wealth_pc * elite_population[i]
        famine_ratio[i] = accum_wealth[i]/wealth_threshold[i]
        if (famine_ratio[i] > 1):
            commoner_population[i+1] = commoner_population[i] + time_step*((commoners_birth_rate*commoner_population[i] - nominal_death_rate*commoner_population[i]))
            elite_population[i+1] = elite_population[i] + time_step*((elites_birth_rate*elite_population[i] - nominal_death_rate*elite_population[i]))
            nature[i+1] = nature[i] + time_step*(((regeneration_rate * nature[i] * (nature_carrying_capacity - nature[i])) - (production_factor * commoner_population[i] * nature[i])))
            accum_wealth[i+1] = accum_wealth[i] + time_step*((production_factor * commoner_population[i] * nature[i]) - subsistence_salary * commoner_population[i] - inequality_factor * subsistence_salary * elite_population[i])
        else:
            commoner_population[i+1] = commoner_population[i] + time_step*((commoners_birth_rate*commoner_population[i] - (nominal_death_rate + (1- famine_ratio[i])* (famine_death_rate - nominal_death_rate)) *commoner_population[i]))
            elite_population[i+1] = elite_population[i] + time_step*((elites_birth_rate*elite_population[i] - (nominal_death_rate + (1- famine_ratio[i]*inequality_factor)* (famine_death_rate - nominal_death_rate)) * elite_population[i]))
            nature[i+1] = nature[i] + time_step*(((regeneration_rate * nature[i] * (nature_carrying_capacity - nature[i])) - (production_factor * commoner_population[i] * nature[i])))
            accum_wealth[i+1] = accum_wealth[i] + time_step* ((production_factor * commoner_population[i] * nature[i]) - famine_ratio[i] * subsistence_salary * commoner_population[i] - famine_ratio[i] * inequality_factor * subsistence_salary * elite_population[i])


########################## CALLING HANDY ##################################################

HANDY(time)
            
########################## PLOTTING THE RESULTS ###########################################

plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
plt.title("POPULATION WITH TIME")
plt.plot(time,commoner_population,"rx", label = "Commoners")
plt.plot(time,elite_population,"bo", label = "Elites")
plt.ylabel("Population", fontweight='bold')
plt.xlabel("Time (years)", fontweight='bold')
plt.legend()

plt.subplot(1,3,2)
plt.title("NATURE WITH TIME")
plt.plot(time,nature,"gx", label = "Nature-Eco-$s")
plt.ylim(bottom = 0)
plt.ylabel("Nature (Eco Dollars - *$)", fontweight='bold')
plt.xlabel("Time (years)", fontweight='bold')
plt.legend()
plt.subplot(1,3,3)
plt.title("WEALTH WITH TIME")
plt.plot(time,accum_wealth, "bx", label = "Wealth-Eco-$s")
plt.ylabel("Wealth (Eco Dollars - *$)", fontweight='bold')
plt.xlabel("Time (years)", fontweight='bold')
plt.legend()
plt.suptitle("COMPARATIVE SIMULATION OF HANDY ABM AND ODE VERSIONS", fontweight='bold')
plt.savefig("L_collapse_unequal.jpg", )
plt.show()
```
