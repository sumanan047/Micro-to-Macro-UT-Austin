```python
# The notebook takes the csv file generated from the netlogo file and generate better plots

import numpy as np   # just in case we need to handle arrays
import matplotlib.pyplot as plt  # for plotting
import pandas as pd   # to convert csv to dataframe


#reading the csvs from netlogo to get dataframes

df_population = pd.read_csv('population_movement.csv',header = 13)
df_population.head()

df_wealth = pd.read_csv('wealth_movement.csv',header = 12)
df_wealth.head()

df_nature = pd.read_csv('nature_movement.csv',header = 12)
df_nature.head()

# Chopping up the required columns

# population of commoners and elites

df_new_pop_mv = df_population.iloc[:,[0,1]]
df_new_pop_mv.head()

try:
    df_new_elite_mv = df_population.iloc[:,[0,5]]
    df_new_elite.head()
except:
    pass

# wealth in the model

df_new_wealth_mv = df_wealth.iloc[:,[0,1]]
#df_new_wealth.head()

# nature in the model

df_new_nature_mv = df_nature.iloc[:,[0,1]]
#df_new_nature.head()
```

```python
########################## PLOTTING THE RESULTS ###########################################

fontname="Times New Roman"
linewidth = 3
parameters = {'axes.labelsize': 15,
          'axes.titlesize': 20}
plt.rcParams.update(parameters)

plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
plt.title("POPULATION WITH TIME",   fontname = fontname)
plt.plot(df_new_pop['x'],df_new_pop['y'],"c-.",linewidth=linewidth,label = "Commoners - Vanilla")
plt.plot(df_new_pop_it['x'],df_new_pop_it['y']* 10,"g:",linewidth=linewidth,label = "Commoners - Intelligent - X10")
plt.plot(df_new_pop_mv['x'],df_new_pop_mv['y']* 10,"b--",linewidth=linewidth,label = "Commoners - Movement - X10")
#plt.plot(df_new_elite['x'],df_new_elite['y.1'],"r:",linewidth=linewidth, label = "Elites - Intelligent")
#plt.plot(df_new_elite_it['x'],df_new_elite_it['y.1'],"r:",linewidth=linewidth, label = "Elites - Intelligent")
#plt.plot(df_new_elite_mv['x'],df_new_elite_mv['y.1'],"r--",linewidth=linewidth, label = "Elites - Movement")
plt.ylim(bottom=0)
plt.xlim(left=0)
plt.tick_params(axis ='both', which ='major',direction ="in", labelsize = 12, pad = 12)
plt.ylabel("Population", fontweight='bold', fontname = fontname)
plt.xlabel("Time (years)", fontweight='bold', fontname = fontname)
plt.legend()

plt.subplot(1,3,2)
plt.title("NATURE WITH TIME")
plt.plot(df_new_nature['x'],df_new_nature['y'],"c-.",linewidth=linewidth, label = "Nature - Eco-$s - Vanilla")
plt.plot(df_new_nature_it['x'],df_new_nature_it['y'],"g:",linewidth=linewidth, label = "Nature - Eco-$s - Intelligent")
plt.plot(df_new_nature_mv['x'],df_new_nature_mv['y'],"b--",linewidth=linewidth, label = "Nature - Eco-$s - Movement")
plt.ylim(bottom=0)
plt.xlim(left=0)
plt.tick_params(axis ='both', which ='major',direction ="in", labelsize = 12, pad = 12)
plt.ylabel("Nature (Eco Dollars - *$)", fontweight='bold', fontname = fontname)
plt.xlabel("Time (years)", fontweight='bold', fontname = fontname)
plt.legend()

plt.subplot(1,3,3)
plt.title("WEALTH WITH TIME")
plt.plot(df_new_wealth['x'],df_new_wealth['y'],"c-.",linewidth=linewidth, label = "Wealth - Eco-$s - Vanilla")
plt.plot(df_new_wealth_it['x'],df_new_wealth_it['y']*10,"g:",linewidth=linewidth, label = "Wealth - Eco-$s - Intelligent - X10")
plt.plot(df_new_wealth_mv['x'],df_new_wealth_mv['y']*10,"b--",linewidth=linewidth, label = "Wealth - Eco-$s - Movement - X10")
plt.ylim(bottom=0)
plt.xlim(left=0)
plt.tick_params(axis ='both', which ='major',direction ="in", labelsize = 12, pad = 12)
plt.ylabel("Wealth (Eco Dollars - *$)", fontweight='bold', fontname = fontname)
plt.xlabel("Time (years)", fontweight='bold', fontname = fontname)
plt.legend()
plt.suptitle("COMPARATIVE SIMULATION OF HANDY ABM VARIANTS", fontweight='bold', fontname = fontname, fontsize= 20)

plt.tight_layout()

plt.savefig("Intelligent_and_random_movement_simulation.jpg")
plt.show()
```
